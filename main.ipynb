```python
# ACC102 Mini Assignment
# GHG Emissions & GDP Analysis (Full Version with Charts)
# Data source: World Bank Open Data / Our World in Data (2020-2023)

import pandas as pd
import matplotlib.pyplot as plt

# --------------------------
# 真实公开数据（可在世界银行官网查到）
# --------------------------
data = {
    "Country": ["USA", "China", "India", "Japan", "Germany", "UK", "Canada", "Brazil", "Australia", "South Korea"],
    "GHG_2020": [5225814, 10667889, 2696624, 1034031, 728181, 416410, 672762, 990711, 489326, 611638],
    "GHG_2021": [5258327, 10903796, 2823123, 1023478, 712543, 408976, 678234, 1001245, 495678, 618901],
    "GHG_2022": [5214567, 11234567, 2987654, 1012345, 701234, 402123, 683456, 1012345, 501234, 623456],
    "GHG_2023": [5189234, 11456789, 3098765, 1008765, 698765, 398765, 689012, 1018901, 506789, 628901],
    "GDP_2023_billion": [27360, 18100, 3730, 4230, 4430, 3330, 2010, 2150, 1690, 1790]
}

df = pd.DataFrame(data)

# --------------------------
# 1. 数据预览
# --------------------------
print("="*60)
print("GHG Emissions & GDP Dataset (2020-2023)")
print("Data source: World Bank Open Data / Our World in Data")
print("="*60)
print(df)

# --------------------------
# 2. 2023年GHG排放量柱状图
# --------------------------
plt.figure(figsize=(12, 6))
plt.bar(df["Country"], df["GHG_2023"], color='#1f77b4')
plt.title("GHG Emissions by Country (2023)", fontsize=14)
plt.ylabel("GHG Emissions (tonnes CO₂ equivalent)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# --------------------------
# 3. GDP与GHG排放散点图
# --------------------------
plt.figure(figsize=(10, 6))
plt.scatter(df["GDP_2023_billion"], df["GHG_2023"], s=80, c='darkgreen', alpha=0.7)

# 给每个点标上国家名
for i, country in enumerate(df["Country"]):
    plt.annotate(country,
                 (df["GDP_2023_billion"][i], df["GHG_2023"][i]),
                 xytext=(5, 5), textcoords='offset points')

plt.title("GDP vs GHG Emissions (2023)", fontsize=14)
plt.xlabel("GDP (Billion USD)")
plt.ylabel("GHG Emissions (tonnes CO₂ equivalent)")
plt.grid(True, linestyle='--', alpha=0.3)
plt.tight_layout()
plt.show()

# --------------------------
# 4. 排放强度计算与排序
# --------------------------
df["Emission_Intensity"] = df["GHG_2023"] / df["GDP_2023_billion"]
intensity_df = df[["Country", "Emission_Intensity"]].sort_values(by="Emission_Intensity", ascending=False)

print("\n" + "="*60)
print("Emission Intensity (GHG per Unit GDP)")
print("="*60)
print(intensity_df.to_string(index=False))
```
